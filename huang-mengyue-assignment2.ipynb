```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

## RGB Analysis Case 1: Spider Woman Jennie
Jennie (my besty) says that she has been so busying commuting between various school buildings this semester, that sometimes she feels like she needs to be a Spider Woman to be on time. I want analyzed how much each color channel are there in the picture.

```python
im1 = Image.open('1.jpg')

im1_data = np.array(im1)
plt.imshow(im1_data)
```

### 1.1 Seperate Graph for RGB Composition

```python
fig, ax = plt.subplots(3,1,figsize=(5,7))

ax[0].hist(im1_data[:,:,0].flatten(),bins=256,color='red')
ax[0].set_xlabel('Red Channel Bins')
ax[0].set_ylabel('Per Pixel each Bin')

ax[1].hist(im1_data[:,:,1].flatten(),bins=256,color='green')
ax[1].set_xlabel('Green Channel Bins')
ax[1].set_ylabel('Per Pixel each Bin')

ax[2].hist(im1_data[:,:,2].flatten(),bins=256,color='blue')
ax[2].set_xlabel('Blue Channel Bins')
ax[2].set_ylabel('Per Pixel each Bin')

plt.show()
```

### 1.2. Combined Graph for RGB Composition

```python
fig,ax = plt.subplots(figsize=(10,3))

ax.hist(im1_data[:,:,0].flatten(), bins=256, color='red',  alpha=0.5)
ax.hist(im1_data[:,:,1].flatten(), bins=256, color='green',alpha=0.5)
ax.hist(im1_data[:,:,2].flatten(), bins=256, color='blue', alpha=0.5)

plt.xlabel('RGB Channel Bins')
plt.ylabel('Per Pixel each Bin')

plt.show()
```
